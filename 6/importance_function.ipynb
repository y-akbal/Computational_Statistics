```python
import scipy
import numpy as np
from scipy.stats import rv_continuous
from matplotlib import pyplot as plt
```

```python
f= lambda x: np.exp(-x)/(1+x**2)
f_uni = lambda x: x/(x+1e-1000)
```

```python
x = np.arange(0,1,0.001)
plt.plot(x, f(x), label = "our function")
plt.plot(x, f_uni(x), label = "uniform_dist")
plt.legend()
plt.show()
```

```python
imp_func = lambda x: np.exp(-x)
f = lambda x: 1/(1+x**2) if 0<=x<=1 else 0
```

```python
x = np.arange(0, 10, 0.0001)
plt.plot(x, imp_func(x), label = "importance_function")
plt.plot(x, [f(x) for x in x], label= "the function")
plt.legend()
plt.show()
```

```python
imp_func = lambda x: 1/((1+x**2)*np.pi)
f = lambda x: np.pi*np.exp(-x) if 0<=x<=1 else 0
```

```python
x = np.arange(-10, 10, 0.0001)
plt.plot(x, imp_func(x), label = "importance_function")
plt.plot(x, [f(x) for x in x], label= "the function")
plt.legend()
plt.show()
```
