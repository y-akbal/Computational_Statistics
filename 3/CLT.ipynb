```python
import scipy
from matplotlib import pyplot as plt
import numpy as np
```

```python
chi2 = scipy.stats.chi2(1)
plt.figure(figsize=(5,3))
plt.hist(chi2.rvs(10000), density = True, bins = 100, label = "chi2 with 1 degrees of freedom");
plt.legend()
plt.show()
```

```python
n = 2  ## Sample size
sample_means = np.array([(chi2.rvs(n).mean() - 1)/((2/n)**0.5) for i in range(10000)])
```

```python
plt.hist(sample_means, bins = 100, density = True, label ="fake");
plt.hist(np.random.randn(10000), bins = 100, density = True, label = "real")
plt.legend()
plt.show()
```
